```python
import pandas as pd
import plotly.express as px


def ler_dados(arquivo):
    x, y = [], []
    with open(arquivo, "r") as f:
        for linha in f:
            if linha.strip():  # Ignora linhas vazias
                partes = linha.split()
                if len(partes) >= 2:
                    x.append(float(partes[0]))
                    y.append(float(partes[1]))
    return x, y
```

```python
x, y = ler_dados(r"D:\programação no windows\introFiscomp\introFiscomp\projeto_5\map.dat")
df = pd.DataFrame({"xi": x, "xi+1": y})

# df = pd.read_csv("projeto_5/map.dat", delimiter="\t", header=None)


fig = px.scatter(x="xi", y="xi+1", data_frame=df, title="Mapa Logístico")


fig.show()
```
